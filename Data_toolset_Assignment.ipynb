Q3. Explain the following terms:

1. The difference in np.array, np.asarray and np.asanyarray.

Answer :

**np.array** - This function creates a new array from the input data.

By default, it copies the input data, which means it always returns a new array.

**np.asarray** - This function converts the input to an array, but it does not copy the input data if input is already an array.

If the input is already a numpy array of same dtype, no copy is performed, which can be more efficient.

**np.asanyarray** - This function is similar tp 'np.asarray', but it passes through subclasses of 'ndarray'.

```python
# example of np.array

import numpy as np
data = [1,2,3]
array1 = np.array(data)
array1
```

```python
# example of np.asarray

import numpy as np
data = [1,2,3]
array2 = np.asarray(data)
print(array2)

existing_array = np.array([4,5,6])
array = np.asarray(existing_array)
print(array is existing_array)
```

Q18. With Bokeh, plot a line chart of sine wave function, add grid lines, label the axes, and set the title as 'Sine wave function'.

```python
import numpy as np
from bokeh.plotting import figure, show, output_notebook
from bokeh.io import push_notebook

# Generate data for the sine wave
x = np.linspace(0, 4 * np.pi, 100)
y = np.sin(x)

# Create a new plot with a title and axis labels
p = figure(title="Sine wave function", x_axis_label='x', y_axis_label='y')

# Add a line renderer with legend and line thickness
p.line(x, y, legend_label="sin(x)", line_width=2)

# Customize grid lines
p.xgrid.grid_line_color = 'gray'
p.ygrid.grid_line_color = 'gray'
p.xgrid.grid_line_alpha = 0.5
p.ygrid.grid_line_alpha = 0.5

# Show the results
show(p, notebook_handle=True)
```

Q19.Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their
values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical
Bar Chart'.

```python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.transform import factor_cmap
from bokeh.palettes import Spectral6

# Enable Bokeh output in Jupyter Notebook (if using Jupyter)
output_notebook()

# Generate random categorical data
categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(1, 100, size=len(categories))

# Create a DataFrame
data = pd.DataFrame({'categories': categories, 'values': values})

# Create a ColumnDataSource
source = ColumnDataSource(data)

# Create the figure
p = figure(x_range=categories, title="Random Categorical Bar Chart",
           x_axis_label='Category', y_axis_label='Value')

# Create the bar chart
p.vbar(x='categories', top='values', width=0.9, source=source,
       legend_field="categories", line_color='white',
       fill_color=factor_cmap('categories', palette=Spectral6, factors=categories))

# Add hover tooltips
hover = HoverTool()
hover.tooltips = [("Category", "@categories"), ("Value", "@values")]
p.add_tools(hover)

# Customize the plot
p.xgrid.grid_line_color = None
p.y_range.start = 0
p.legend.orientation = "horizontal"
p.legend.location = "top_center"

# Show the plot
show(p)
```

Q20 . Using Plotly, create a basic line plot of a random generated dataset, label the axes, and set the title as
'Simple Line Plot'

```python
import numpy as np
import plotly.graph_objects as go

x = np.arange(100)
y = np.random.randn(100)

fig = go.Figure()
fig.add_trace(go.Scatter(x=x, y=y, mode='lines', name='Random Data'))

fig.update_layout(
    title = 'Simple Line Plot',
    xaxis_title = 'X Axis',
    yaxis_title = 'Y Axis'
)
fig.show()
```

Q21. Using Plotly, create an interactive pie chart of randomly generated data, add labels and percentages, set
the title as 'Interactive Pie Chart'.

```python
import numpy as np
import plotly.graph_objects as go

labels = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(1,100, size=len(labels))

fig = go.Figure(data=[go.Pie(labels=labels, values=values, textinfo='label+percent', hole=0.3)])

fig.update_layout(title='Interactive Pie Chart')
fig.show()
```
