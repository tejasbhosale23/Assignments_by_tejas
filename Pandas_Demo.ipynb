```python
A = int(input('enter the starting num'))
B = int(input('enter the last num'))
M = int(input('enter the divisible num'))
def sumDivisibles(A, B, M):
    sum = 0
    for i in range(A, B + 1):
        if (i % M == 0):
            sum += i
    return sum

def sum1divisible(A,B,M):
    l = []
    for i in range(A, B + 1):
        if (i % M == 0):
            l.append(i)
    return l
if __name__=="__main__":
    print(sumDivisibles(A, B, M))
    print(sum1divisible(A,B,M))
```
